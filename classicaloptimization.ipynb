```python
import networkx as nx
import time
start_time = time.time()
# Create a graph
G = nx.Graph()

# Input number of nodes and nodes' names
num_nodes = int(input("Enter number of nodes: "))
nodes = []
for i in range(num_nodes):
    node_name = input(f"Enter name of node {i+1}: ")
    nodes.append(node_name)
    G.add_node(node_name)

# Input edges with distances
print("Enter edges (node1 node2 distance):")
num_edges = int(input("Enter number of edges: "))
for _ in range(num_edges):
    n1, n2, dist = input().split()
    dist = float(dist)
    G.add_edge(n1, n2, weight=dist)

# Input source and target for path search
source = input("Enter source node: ")
target = input("Enter target node: ")

# Use Dijkstra's algorithm to find shortest path and distance
try:
    shortest_path = nx.dijkstra_path(G, source, target, weight='weight')
    shortest_distance = nx.dijkstra_path_length(G, source, target, weight='weight')

    print(f"Shortest path: {' -> '.join(shortest_path)}")
    print(f"Total distance: {shortest_distance}")
except nx.NetworkXNoPath:
    print(f"No path found between {source} and {target}")
except Exception as e:
    print("Error:", e)

end_time = time.time()
execution_time = end_time - start_time

print(f"Execution time for Dijkstra: {execution_time:.6f} seconds")
```
